# JSON : 실시간 지하철 위치

1. 나의 인증키,타입, 서비스, URL 등 요청인자(변수명) 정의 
2. 기존에 제시한 URL을 f스트링으로 변수를 넣을 수 있게 변환해줌 
3. requests를 import하고, get함수로 URL을 불러와줌 -> response라는 변수에 넣어줌
4. response함수를 json으로 읽어줌 -> json_data라는 변수에 넣어줌
5. json_data의 key값을 조회하여 확인함. 불러와야할 대상을 명료하게 좁히는 단계임
6. key값이 되는 값을 json_data['  ']로 정의해주고, data라는 변수에 넣어줌
7. pandas를 임포트하여 data를 DataFrame형태로 만들어줌

```python
# 기본적인 정보 입력해 

API_KEY = '55674272676d65793832626b4f6d75'

TYPE = 'json'
SERVICE = 'realtimePosition'
START_INDEX = '0'
END_INDEX = '5'
subwayNm = '1호선'
URL = f'http://swopenAPI.seoul.go.kr/api/subway/{API_KEY}/{TYPE}/{SERVICE}/{START_INDEX}/{END_INDEX}/{subwayNm}'
# http://swopenAPI.seoul.go.kr/api/subway/(인증키)/xml/realtimePosition/0/5/1호선
```

```python
# URL 읽어와
import requests
response = requests.get(URL)
```

```python
#json형식으로 가져와
json_data = response.json()
json_data
```

```python
#key값 확인해
json_data.keys()
```

```python
#키값 불러와서 data에 넣어
data = json_data['realtimePositionList']
```
